```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('eda_outlier.csv')
```

```python
df.info()
```

```python
df.drop(columns={'facing'},inplace=True)
```

```python
df[df['property_type'] == 'house']['floor'].median()
```

```python
df['floor'].fillna(2.0,inplace=True)
```

```python
df.age.value_counts()
```

```python
df[df['age'] == 'undefined']
```

```python
def mode_based_imputation(row):
    if row['age'] == 'undefined':
        mode_value = df[(df['sector'] == row['sector']) & (df['property_type'] == row['property_type'])]['age'].mode()
        # If mode_value is empty (no mode found), return NaN, otherwise return the mode
        if not mode_value.empty:
            return mode_value.iloc[0] 
        else:
            return np.nan
    else:
        return row['age']
```

```python
df['age'] = df.apply(mode_based_imputation,axis=1)
```

```python
df.age.value_counts()
```
